```python
df.sort_values(by='Sales', ascending=False, inplace=True)

df
```

```python
#Calculate the total profit for each region.


df = df.groupby('Region').agg({'Profit': 'sum'}).reset_index()
```

```python
df = df.groupby('Product').agg({'Quantity': 'mean'}).reset_index()
```

```python
df
```

```python
def max_sales(x):
    a = x.max()
    if a > 500:
        return a
    else:
        return np.nan
    

df = (df.groupby('Region').agg({'Sales': max_sales}).dropna().reset_index())
```

```python
df
```

```python
#find 2nd highest sales


def sec_max(x):
    a=x.max()
    unique= x.drop_duplicates().sort_values(ascending=False)
    
    if len(unique)>=2:
        return unique.iloc[1]
    else:
        return np.nan
       
        
df.groupby('Region').agg({'Sales': sec_max}).reset_index().dropna()
```

```python
#Total sales and total profit for the entire dataset.
df[['Sales', 'Profit']].sum()
```

```python
#total sale as per region

def total_sum(x):
    return x.sum()



df.groupby('Region').agg(total_sale=('Sales',total_sum)).reset_index()
```

```python
#Average profit per product

def avg(x):
  return x.mean()


df =df.groupby('Product').agg(Average_Profit=('Profit',avg)).reset_index()
  
```

```python
#Find the second highest Sales per Region

def sec_sales(x):
    unique= x.drop_duplicates().sort_values(ascending=False)
    
    if len(unique)>=2:
        return unique.iloc[1]
    else:
        return np.nan
    
df.groupby('Region').agg(sec_highest=('Sales',sec_sales)).reset_index().dropna()
        
```
