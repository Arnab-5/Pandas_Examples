```python


# First DataFrame
df1 = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104],
    'Name': ['Amit', 'Neha', 'Ravi', 'Pooja']
})

# Second DataFrame
df2 = pd.DataFrame({
    'Emp_ID': [102, 103, 104, 105],
    'Department': ['HR', 'IT', 'Finance', 'Marketing']
})

# Merge DataFrames using common column
merged_df = pd.merge(df1, df2, on='Emp_ID')

print(merged_df)
```

```python
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Second DataFrame (different columns)
df2 = pd.DataFrame({
    'C': [7, 8, 9],
    'D': [10, 11, 12]
})

# 1️⃣ Concatenate along rows (axis=0)
row_concat = pd.concat([df1, df2], axis=0)
print("Row-wise concatenation:\n", row_concat)

# 2️⃣ Concatenate along columns (axis=1)
col_concat = pd.concat([df1, df2], axis=1)
print("\nColumn-wise concatenation:\n", col_concat)
```

```python
import pandas as pd
import numpy as np

# Create a date range (datetime index)
date_index = pd.date_range(start='2021-01-01',end='2021-12-31' , freq='D')

# Create DataFrame with random integers
df = pd.DataFrame({
    'Value': np.random.randint(10, 100, size=len(date_index))
}, index=date_index)

print("Original DataFrame:\n", df.head())

# Resample to monthly frequency and compute mean
monthly_mean = df.resample('W').mean()

print("\nMonthly Mean:\n", monthly_mean)
```

```python
import pandas as pd
import numpy as np

# Create MultiIndex
index = pd.MultiIndex.from_tuples(
    [
        ('Electronics', 'Mobile'),
        ('Electronics', 'Laptop'),
        ('Furniture', 'Chair'),
        ('Furniture', 'Table')
    ],
    names=['Category', 'SubCategory']
)

# Create DataFrame
df = pd.DataFrame(
    {
        'Sales': np.random.randint(100, 500, 4),
        'Profit': np.random.randint(20, 100, 4)
    },
    index=index
)

print(df)

```
