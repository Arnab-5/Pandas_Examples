```python
#1. Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float). Add some data and sort the array by age.
import numpy as np

# Define data type for structured array
dtype = [('name', 'U10'), ('age', 'i4'), ('weight', 'f4')]

# Create structured array with some data
data = np.array([
    ('Alice', 25, 55.5),
    ('Bob', 22, 70.2),
    ('Charlie', 30, 68.0),
    ('David', 28, 75.0)
], dtype=dtype)

# Sort array by 'age'
sorted_data = np.sort(data, order='age')

print("Original array:\n", data)
print("\nSorted by age:\n", sorted_data)
```

```python
import numpy as np

# Define structured array with 'x' and 'y' as integers
dtype = [('x', 'i4'), ('y', 'i4')]

# Create structured array with some points
points = np.array([
    (1, 2),
    (4, 6),
    (7, 8)
], dtype=dtype)

# Extract x and y as regular arrays
x = points['x']
y = points['y']

# Compute Euclidean distance between each pair
n = len(points)
distances = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        distances[i, j] = np.sqrt((x[i] - x[j])**2 + (y[i] - y[j])**2)

print("Points:\n", points)
print("\nEuclidean distance matrix:\n", distances)
```
