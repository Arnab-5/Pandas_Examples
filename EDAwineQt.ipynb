6. Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

mean_val = df['quality'].mean()
median_val = df['quality'].median()
mode_val = df['quality'].mode()[0]

plt.figure(figsize=(8,5))
sns.histplot(df['quality'], bins=6)

plt.axvline(mean_val, linestyle='--', label=f'mean:{mean_val:2f}')
plt.axvline(median_val, linestyle='-', label=f'Median: {median_val}')
plt.axvline(mode_val, linestyle=':', label=f'Mode: {mode_val}')

plt.xlabel('Quality')
plt.ylabel('Quantity')
plt.title('Central Tendency of Quality')
plt.legend()
plt.show()
```
